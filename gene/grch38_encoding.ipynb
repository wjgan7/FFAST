```python
from Bio import Seq
import numpy as np
```

```python
last_names = ['chrX', 'chrY', 'mito']
def seq_name(i):
    assert i >= 0 and i < 25
    if i < 22:
        return f'chr{i+1}'
    else:
        return last_names[i-22]
```

```python
seqs = []
for i in range(25):
    text_file = open(f'grch38_dna/{seq_name(i)}.txt', 'r')
    seqs.append(next(text_file))
    text_file.close()
```

```python
for seq in seqs:
    print(seq[:30])
```

```python
for seq in seqs:
    cnts = {}
    for symbol in seq:
        if symbol not in cnts:
            cnts[symbol] = 0
        cnts[symbol] += 1
    print(cnts)
```

[This NCBI link](https://blast.ncbi.nlm.nih.gov/Blast.cgi?CMD=Web&PAGE_TYPE=BlastDocs&DOC_TYPE=BlastHelp) offers a little bit of explanation on what the codes mean. I think the gist is that N means unknown, A C G T are your standard adenine, cytosine, guanine, and thymine, a, c, g, t are also your standard but put in lower case to indicate they are within a repetitive pattern. There are a few Y, R, M, K. Those represent that the nucleotide can either be A or C, A or T, or something of that sort.

I think we can just convert the lower cases to upper case and replace the Y, R, M, K, etc with N. I am a bit unsure about what to do with the Ns because they are pretty important yet seem to come in long continguous sequences (like the beginning of chromosomes) and so I think they might skew how the cross correlation looks. I'll try encode two versions.

```python
code_with_N = {
        'a': [-1, -1, -1],
        'A': [-1, -1, -1],
        'c': [-1, -1, 1],
        'C': [-1, -1, 1],
        'g': [-1, 1, -1],
        'G': [-1, 1, -1],
        't': [-1, 1, 1],
        'T': [-1, 1, 1],
        'N': [1, -1, -1]}
code_without_N = {
        'a': [-1, -1],
        'A': [-1, -1],
        'c': [-1, 1],
        'C': [-1, 1],
        'g': [1, -1],
        'G': [1, -1],
        't': [1, 1],
        'T': [1, 1]}
```

```python
def calculate_splits(seq):
    splits = []
    split_class = []
    split_sz, curr_isnt_N = 1, seq[0] in code_without_N 
    
    for i in range(1, len(seq)):
        prev_isnt_N, curr_isnt_N = curr_isnt_N, seq[i] in code_without_N
        if not prev_isnt_N == curr_isnt_N:
            splits.append(split_sz / len(seq))
            split_class.append(prev_isnt_N)
            split_sz = 1
        else:
            split_sz += 1
            
    splits.append(split_sz / len(seq))
    split_class.append(curr_isnt_N)
    
    return (splits, split_classes)
```

```python
splits_tuples = []
for seq in seqs:
    splits_tuple = calculate_splits(seq)
    splits_tuples.append(split_tuple)
```

```python
for splits, split_classes in splits_tuples:
    print(len(splits))
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_splits(splits_tuple, name):
    # Code taken from http://peterspangler.com/a-single-stacked-bar-chart-in-matplotlib/
    
    plt.rcParams["figure.figsize"] = [10, 3.5]
    fig, chart_ax = plt.subplots()
    plt.rcdefaults()
    
    y_pos = [0]
    width = 0.05
    
    chart_ax.barh(y_pos, 0, 1.0, align='center', color='white', ecolor='black', label=None)
    
    chart_ax.barh(y_pos, 1, 0.12, align='center', color='#D9DCDE', left=0)
            
    left_pos = 0
    
    for split, split_class in zip(splits_tuple[0], splits_tuple[1]):
        chart_ax.barh(y_pos, [split], width, align='center',
                      color='green' if split_class else 'red',
                      left=left_pos, edgecolor=['black', 'black'], linewidth=0.5)
        left_pos += split
        
    chart_ax.set_yticks([1])
    chart_ax.invert_yaxis()
    chart_ax.set_title(f'{name} (Green is A, C, G, or T (or lowercase) and Red is N)')
    plt.tight_layout()

    plt.savefig(f'grch38_symbol_plots/{name}.png')
```

```python
for i in range(25):
    plot_splits(split_tuples[i], seq_name(i))
```

```python
def encode_with_N(seq):
    vec = np.zeros(3 * len(seq), dtype=np.int8)
    for i, symbol in enumerate(seq):
        if symbol not in code_with_N:
            vec[3*i:3*(i+1)] = code_with_N['N']
        else:
            vec[3*i:3*(i+1)] = code_with_N[symbol]
    return vec
```

```python
def encode_without_N(seq):
    vec = np.zeros(2 * len(seq), dtype=np.int8)
    for i, symbol in enumerate(seq):
        if symbol not in code_without_N:
            continue
        else:
            vec[2*i:2*(i+1)] = code_without_N[symbol]
    return vec
```

```python
test_with_N = encode_with_N(seqs[24])
print(len(test_with_N))
print(test_with_N[:30])
test_without_N = encode_without_N(seqs[24])
print(len(test_without_N))
print(test_without_N[:30])
```
