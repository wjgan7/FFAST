```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy import signal
```

```python
last_names = ['chrX', 'chrY', 'mito']
def seq_name(i):
    assert i >= 0 and i < 25
    if i < 22:
        return f'chr{i+1}'
    else:
        return last_names[i-22]
```

```python
def plot_cross_correlation(cc, file_name):
    plt.figure()
    plt.stem(cc)
    plt.savefig(file_name)
```

```python
def read_code(file_name):
    file = open(file_name, 'r')
    N = int(next(file))
    x = np.zeros(N, dtype=np.int64)
    for (i, line) in enumerate(file):
        x[i] = int(line)
    return x
```

```python
LENGTHS = [50, 100, 500, 1000, 5000, 10000, 50000, 100000]
```

```python
LENGTHS = [50, 100, 500, 1000, 5000, 10000, 50000, 100000]
TRIALS_PER_SEQ = 1

for i in range(24, 25):
    name = seq_name(i)
    
    x = read_code(f'grch38_cc_with_N/{name}.txt')
    for L in LENGTHS:
        print(f'length: {L}')
        if L >= len(x):
            continue
        for k in range(5):
            print(f'trial:{k}')
            start = np.random.randint(0, len(x)-L+1)
            y = x[start : start+L]
            print('about to correlate')
            z = signal.correlate(x, y, mode='valid')
            print('about to plot')
            plot_cross_correlation(z, f'grch38_cc_with_N/{L}/{name}_trial_{k+1}.txt')
            
    x = read_code(f'grch38_cc_without_N/{name}.txt')
    for L in LENGTHS:
        if L >= len(x):
            continue
        for k in range(5):
            start = np.random.randint(0, len(x)-L+1)
            y = x[start : start+L]
            z = signal.correlate(x, y, mode='valid')
            plot_cross_correlation(z, f'grch38_cc_without_N/{L}/{name}_trial_{k+1}.jpgb')
```

```python
plot_cross_correlation(z, f'grch38_cc_plots_without_N/{L}/{name}_trial_{k+1}.jpg')
```
